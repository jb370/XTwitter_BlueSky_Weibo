```python
#Creating the function
def country_sentiment(data, country):
    empty_list = []
    for i in range(0, len(data)):
        text = data["Text"][i]
        positionCountry = text.find(country)
        if (positionCountry == -1):
            empty_list.append(0)
        else:
            empty_list.append(1)
    data["country_yes"] = empty_list
    condition = data['country_yes'] == 1
    data2 = data[condition]
    data3 = data2.reset_index(drop = True)
    sentimentCountry= int(data3["Sentiment"].sum()) / len(data3)
    return sentimentCountry
```
